# Neural Networks with Tensorflow/Keras

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
```

# Loading and Preprocessing data

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Normalization
x_train = x_train.astype('float32')/255.0
x_test = x_test.astype('float32')/255.0
```

```python
#Flattening
x_train = x_train.reshape(-1, 28 * 28) #each image size 28*28 //784
x_test = x_test.reshape(-1, 28*28)
```

```python
#Encoding
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, num_classes=10) #the length of the vector equals the number of classes
y_test = to_categorical(y_test, num_classes=10)
y_train = y_train.reshape(-1, 10)
```

# Neural Network Designation Architecture

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
FFNN = Sequential(
    [
        Dense(128,activation='relu', input_shape=(784,)), #input layer
        Dropout(0.2), #dropout for regularization
        Dense(64, activation='relu'), #hidden layer
        Dense(10, activation='softmax') #output layer for 10 classes
    ]
)
```

```python
FFNN.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
FFNN.summary()
```

# Model Training with backward propagation

```python
model = FFNN.fit(
    x_train, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=32,
    verbose=2
)
```

# Model Evaluation and Visualization

```python
loss, accuracy = FFNN.evaluate(x_test, y_test, verbose=2)
print(f'Test Loss: {loss*100:.2f}')
print(f'Test Accuracy: {accuracy*100:.2f}')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(model.history['loss'], label='Training Loss')
plt.plot(model.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation loss')
plt.legend()
plt.show()
```

```python
plt.plot(model.history['accuracy'], label='Training Accuracy')
plt.plot(model.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation accuracy')
plt.legend()
plt.show()
```
