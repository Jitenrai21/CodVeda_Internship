# Support Vector Machine (SVM) Classification

```python
import pandas as pd
dataset = pd.read_csv('Placement1.csv')
dataset.head(3)
```

```python
dataset.drop(columns=['Student_ID'], inplace=True)
dataset.head(3)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x = dataset.iloc[:,:-1]
y = dataset['Placement']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# Model Training

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(x_train, y_train)
```

```python
# For Kernel = linear
svc_lin = SVC(kernel='linear')
svc_lin.fit(x_train, y_train)
```

```python
# For Kernel = rbf
svc_rbf = SVC(kernel='rbf')
svc_rbf.fit(x_train, y_train)
```

# Performance Comparision

```python
from sklearn.metrics import precision_score, recall_score, f1_score
y_pred_lin = svc_lin.predict(x_test)
ps_lin = precision_score(y_test, y_pred_lin)
rs_lin = recall_score(y_test, y_pred_lin)
f1_lin = f1_score(y_test, y_pred_lin)

y_pred_rbf = svc_rbf.predict(x_test)
ps_rbf = precision_score(y_test, y_pred_rbf)
rs_rbf = recall_score(y_test, y_pred_rbf)
f1_rbf = f1_score(y_test, y_pred_rbf)
```

```python
print('The performance metrices scores for SVM trained in kernal Linear are:')
print(f'Accuracy score: {svc_lin.score(x_test, y_test)*100}')
print(f'Precision score: {ps_lin*100}')
print(f'Recall score: {rs_lin*100}')
print(f'F1 score: {f1_lin*100}')
```

```python
print('The performance metrices scores for SVM trained in kernal RBF are:')
print(f'Accuracy score: {svc_rbf.score(x_test, y_test)*100}')
print(f'Precision score: {ps_rbf*100}')
print(f'Recall score: {rs_rbf*100}')
print(f'F1 score: {f1_rbf*100}')
```

# Decision Boundary Visualization

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=svc)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=svc_lin)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=svc_rbf)
plt.show()
```

# Model Evaluation

```python
y_pred = svc.predict(x_test)
ps = precision_score(y_test, y_pred)
rs = recall_score(y_test,y_pred)
f1 = f1_score(y_test, y_pred)
print('The performance metrices scores for SVM are:')
print(f'Accuracy score: {svc.score(x_test, y_test)*100}')
print(f'Precision score: {ps*100}')
print(f'Recall score: {rs*100}')
print(f'F1 score: {rs*100}')
```

```python
from sklearn.metrics import roc_auc_score
y_scores = svc.decision_function(x_test)
auc = roc_auc_score(y_test, y_scores)
print(f'AUC: {auc*100:.2f}')
```

```python
# AUC for more accurate model trained in kernel linear
y_scores_lin = svc_lin.decision_function(x_test)
auc_lin = roc_auc_score(y_test, y_scores_lin)
print(f'AUC: {auc_lin*100:.2f}')
```
