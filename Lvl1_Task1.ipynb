```python
import pandas as pd
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head()
```

# Handling Missing Data

```python
dataset.isnull().sum() 
#this shows the number of null values for each attributes
```

```python
dataset.shape
```

```python
new_data = pd.DataFrame(dataset)
new_data.dropna(subset=['Gender'], inplace=True)
new_data.isnull().sum()
```

```python
new_data.shape
```

```python
mean = dataset['LoanAmount'].mean()
dataset['LoanAmount'].fillna(mean, inplace= True)
print(mean)
dataset.head()
```

# Label Encoding of Categorical Data

```python
dataset['Property_Area'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
la = LabelEncoder()
dataset['Property_Area'] = la.fit_transform(dataset['Property_Area'])
print(dataset['Property_Area'])
```

# Normalization

```python
import seaborn as sns
import matplotlib as plt
```

```python
dataset.describe()
```

```python
sns.displot(dataset['CoapplicantIncome']) # the difference between the values is very high
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome_ss'] = ss.transform(dataset[['CoapplicantIncome']])
dataset['CoapplicantIncome_ss'].describe()
```

```python
sns.displot(dataset['CoapplicantIncome_ss'])
```
