```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('2) Stock Prices Data Set.csv')
data.head()
```

```python
data.isnull().sum()
```

# Preprocessing

```python
high_mean = data['high'].mean()
low_mean = data['low'].mean()
data['high'].fillna(high_mean, inplace=True)
data['low'].fillna(low_mean, inplace=True)
data.isnull().sum()
```

```python
data.drop(columns=['symbol', 'date', 'open','close','volume'], inplace=True)
data.head()
```

```python
data.corr()
```

# Train Test Split

```python
x = data[['high']]
y = data[['low']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x)
sns.scatterplot(x='high', y='low', data= data)
plt.plot(data['high'], y_pred, c='red')
plt.legend(['original data', 'prediction'])
plt.show()
```

# Model Coefficients Interpretation

```python
lr.coef_
```

```python
lr.intercept_
```

# Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
test_pred = lr.predict(x_test)
r2 = r2_score(y_test, test_pred)
mse = mean_squared_error(y_test, test_pred)
print(f"R-squared error: {r2}")
print(f'Mean Squared Error: {mse}')
```
