# Decision Tree Classification

```python
import pandas as pd
dataset = pd.read_csv('1) iris.csv')
dataset.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x = dataset.iloc[:,:-1]
y = dataset['species']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

# Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

# Decision Tree Visualization

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plot_tree(dt)
plt.show()
```

# Pruning to prevent overfitting

```python
dt1 = DecisionTreeClassifier(max_depth=3)
dt1.fit(x_train, y_train)
```

```python
plot_tree(dt1)
plt.show()
```

# Model Evaluation

```python
from sklearn.metrics import f1_score
```

```python
print("Model Accuracy:",dt.score(x_test, y_test)*100)
print("F1 score:",f1_score(y_test, dt.predict(x_test), average='micro')*100)
```
